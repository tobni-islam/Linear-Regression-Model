```python
x = x.to_numpy()
x_train = x[:800,:]
x_test = x[800:,:]
```

```python
y = y.to_numpy()
y_train = y[:800]
y_test = y[800:]
```

```python
w = np.random.random(5) * 10
b = 1000
w,b
```

```python
def predict(w,x,b):
    return np.dot(w,x) + b
```

```python
def loss_func(x,y,w,b):
    return predict(w,x,b) - y
```

```python
def cost_func(x,y,w,b):
    m = x.shape[0]
    s = 0
    for i in range(m):
        s += (loss_func(x[i],y[i],w,b)) ** 2
    return s / (2 * m)
```

```python
print(cost_func(x_train,y_train,w,b))
```

```python
def gradient_func(x,y,w,b):
    m,n = x.shape
    dj_dw = np.zeros(n,)
    dj_db = 0
    for i in range(m):
        for j in range(n):
            dj_dw[j] += loss_func(x[i],y[i],w,b) * x[i,j]
        dj_db += loss_func(x[i],y[i],w,b)
    dj_dw = dj_dw / m
    dj_db = dj_db / m
    return dj_dw,dj_db
```

```python
tmp_dj_dw , tmp_dj_db = gradient_func(x_train,y_train,w,b)
print(f"dj_dw init: {tmp_dj_dw}")
print(f"dj_db init: {tmp_dj_db}")
```

```python
def gradient_descent(x,y,w_init,b_init,alpha,num_iter):
    w = w_init
    b = b_init
    
    for i in range(num_iter):
        dj_dw, dj_db = gradient_func(x_train,y_train,w,b)
        w -= alpha * dj_dw
        b -= alpha * dj_db
        if i % 100 == 0:
            print(f"Iteration {i} / Cost function J = {cost_func(x_train,y_train,w,b)}")
            print(f"w = {w}")
            print(f"b = {b}")
    return w,b
```

```python
m,n = x_train.shape
w = np.zeros(n,)
b = 0
alpha = 0.2
num_iter = 10000

w_final , b_final = gradient_descent(x_train,y_train,w,b,alpha,num_iter)
```

```python
print(f"w  = {w_final}")
print(f"b = {b_final}")
```

```python
#Model Testing
n = y_test.shape[0]
mean_err = 0
for i in range(n):
    mean_err += abs(predict(w_final,x_test[i],b_final) - y_test[i]) / n

print(f"mean error = {mean_err}")
```
